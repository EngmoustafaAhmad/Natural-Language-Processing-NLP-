```python
import spacy

nlp = spacy.load("en_core_web_sm")
nlp.pipe_names
```

```python
text = "tesla inc is going to acquire twitter for 45$ billion"
doc = nlp(text)

print(doc.ents)
for ent in doc.ents:
    print(ent.text,"|",ent.label_,"|",spacy.explain(ent.label))
```

```python
from spacy import displacy
displacy.render(doc,style="ent")
```
