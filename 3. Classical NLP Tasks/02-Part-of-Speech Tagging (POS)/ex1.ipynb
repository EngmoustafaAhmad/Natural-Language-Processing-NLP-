```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("""Revenue was $51.7 billion and increased 20%
Operating income was $22.2 billion and increased 24%
Net income was $18.8 billion and increased 21%
Diluted earnings per share was $2.48 and increased 22%
Digital technology is the most malleable resource at the world's disposal to overcome constraints
Solid commercial execution, represented by strong bookings growth driven by long-term Azure commit """)

filtered_words = []

for word in doc:
    if word.pos_ not in ["SPACE","PUNCT","X"]:
        filtered_words.append(word)

for word in filtered_words:
    print(
        word.text,
        "=>",
        word.pos_,
        spacy.explain(word.pos_),
        word.tag_,
        spacy.explain(word.tag_),
    )
```
